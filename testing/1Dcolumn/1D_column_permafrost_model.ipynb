### 1D Columnar Permafrost Model

This problem demonstrates a typical 1D column series of runs, from spinup to full run (i.e. one member of an ensemble).  

This is a multi-stage process, but we include a few unnecessary steps here to test and demonstrate.  This demo shows a real example of how ATS is used to build up a full simulation suite, including intermediate steps.  For instance, we show multiple runs of the same problem, with slightly different physical representations, to demonstrate how we incrementally add complexity as a testing mechanism.

This demo also is useful for seeing how layering works in ATS, as there are heterogeneous properties throughout (peat and mineral soil layers).

The basic layout of this sequence is as follows:

* test0: Run a steady-state problem to establish a water table
* test1: Rain on the steady-state solution (adds surface ponding to test0)
* test2: Freeze the steady-state water table (adds thermal to test0)
* test4: Freeze/thaw the subsurface by applying cyclic boundary conditions (transient problem that is somewhat typical of what a real problem will look like)
* test7: Full run (adds SEB, surface ponding/energy)

```python
%matplotlib inline
import sys,os
sys.path.append(os.path.join(os.environ['ATS_SRC_DIR'],'tools', 'utils'))
import parse_xmf
import column_data
import colors
import numpy as np
import matplotlib.cm
from matplotlib import pyplot as plt
```

#### test0: Richards steady-state

Solve the steady-state Richards equation to establish a water table in hydrostatic equilibrium.

Two equivalent runs are done -- the solutions should be identical, but the second adds a density and viscosity model.

* test0-i: no temperature, constant (incompressible) density and constant viscosity
* test0-ii: temperature prescribed at constant value

#### test1: rain on steady state

Couple surface and subsurface water, raining on the column until the column is full and ponded water expresses on the surface.

Three equivalent runs are done:
* test1-i: From test0-i.  Again, this run includes no temperature.
* test1-ii: From test0-ii. Isothermal run with prescribed, constant, uniform temperature.  Adds viscosity and density models.
* test1-iii: Fully coupled surface/subsurface flow/energy.  Temperature is initially constant, and no energy is added, so the temperature should stay constant.  

_*This is not a necessary step, and is just done for testing!*_

```python
# plot the solution

def plot(directory, axs, cmap, style):
    # load the data
    dati = column_data.column_data(["pressure", "saturation_liquid"],
                                  directory=directory)
    keys, times, dat = parse_xmf.readATS(directory=directory)
    skeys, stimes, sdat = parse_xmf.readATS(directory=directory, base="visdump_surface_data.h5")
    
    # convert times to days
    times = [time*365.25 for time in times]
    cmapper = colors.cm_mapper(-dati.shape[1]/2,dati.shape[1]-1, cmap=cmap)
    
    # plot saturation of liquid, pressure
    z = dati[0,0,:]
    for i,t in enumerate(times):
        axs[1].plot(dati[1,i,:],z, style, color=cmapper(i))
        axs[0].plot(dati[2,i,:],z, style, color=cmapper(i))

    # inset -- plot surface water
    pd = parse_xmf.getSurfaceData(keys, sdat, 'ponded_depth.cell.0')
    axs[-1].plot(times, pd, style, color=cmapper(dati.shape[1]/2))
    if abs(pd.max()) > 0:
        axs[-1].set_ylim([-.1*pd.max(), pd.max()])

            

fig, axs = column_data.getFigs(True, False)
di = "test1/test1-i.regression.gold"
dii = "test1/test1-ii.regression.gold"
diii = "test1/test1-iii.regression.gold"

plot(di, axs, matplotlib.cm.Blues, '-')
plot(dii, axs, matplotlib.cm.Greens, '--')
plot(dii, axs, matplotlib.cm.Reds, '-.')

axs[0].set_xlim([-.1,1.1])
axs[0].set_xlabel('saturation [-]')
axs[0].set_ylabel('z-coordinate [m]')
axs[1].set_xlabel('pressure [Pa]')
axs[1].set_ylabel('z-coordinate [m]')
axs[1].ticklabel_format(style='sci', axis='x', scilimits=(0,0))

axs[-1].set_xlabel('time [d]')
axs[-1].set_ylabel('ponded depth [m]')
axs[-1].set_xticks([0,100])


# and the zoomed version
fig, axs = column_data.getFigs(True, False)
di = "test1/test1-i.regression.gold"
dii = "test1/test1-ii.regression.gold"
diii = "test1/test1-iii.regression.gold"

plot(di, axs, matplotlib.cm.Blues, '-')
plot(dii, axs, matplotlib.cm.Greens, '--')
plot(dii, axs, matplotlib.cm.Reds, '-.')

axs[0].set_xlim([-.1,1.1])
axs[0].set_xlabel('saturation [-]')
axs[0].set_ylabel('z-coordinate [m]')
axs[1].set_xlabel('pressure [Pa]')
axs[1].set_ylabel('z-coordinate [m]')
axs[1].ticklabel_format(style='sci', axis='x', scilimits=(0,0))
axs[0].set_ylim([4.5,5])
axs[1].set_ylim([4.5,5])
axs[1].set_xlim([0.4e5,1.5e5])

axs[-1].set_xlabel('time [d]')
axs[-1].set_ylabel('ponded depth [m]')
axs[-1].set_xticks([0,100])

plt.show()
```

#### test2: Freeze steady-state to generate ice column

Transient simulations are initialized in the winter to make spinup simpler.  Therefore we start from a fully frozen domain (instead of the partially frozen domain that would be required for starting any other time of year).  

To establish this fully-frozen domain, we freeze the hydrostatic water column from below until the solution reaches steady-state again.  This "pseudo-transient" approach is easier for solvers than solving the steady-state problem.

The bottom boundary temperature is set at -9 F, which is, according to deep borehole data, a reasonable temperature for deep permafrost in Barrow.

NOTE: From here on, transient problems are plotted such that early times have low saturation, and late times are high saturation.

```python
# plot the solution

def plot(directory, axs, style):
    # load the data
    dati = column_data.column_data(["pressure", "saturation_liquid", 
                                    "saturation_ice", "saturation_gas", "temperature"],
                                    directory=directory)
    keys, times, dat = parse_xmf.readATS(directory=directory)

    ice = colors.cm_mapper(-10, times[-1], colors.ice_cmap())
    water = colors.cm_mapper(-10, times[-1], colors.water_cmap())
    gas = colors.cm_mapper(-10, times[-1], colors.gas_cmap())    
    
    # plot saturation of liquid, pressure
    z = dati[0,0,:]
    for i,t in enumerate(times):
        axs[2].plot(dati[5,i,:],z, style, color=gas(t))
        axs[1].plot(dati[1,i,:],z, style, color=water(t))
        axs[0].plot(dati[2,i,:],z, style, color=water(t))
        axs[0].plot(dati[3,i,:],z, style, color=ice(t))
        axs[0].plot(dati[4,i,:],z, style, color=gas(t))
        
fig, axs = column_data.getFigs(False, True)
diii = "test2/test2-iii.regression.gold"

plot(diii, axs, '-')

axs[0].set_xlim([-.1,1.1])
axs[0].set_xlabel('saturation [-]')
axs[0].set_ylabel('z-coordinate [m]')
axs[1].set_xlabel('pressure [Pa]')
axs[2].set_xlabel('temperature [K]')
axs[1].ticklabel_format(style='sci', axis='x', scilimits=(0,0))

# and the zoomed version
fig, axs = column_data.getFigs(False, True)


plot(diii, axs, '-')


axs[0].set_xlim([-.1,1.1])
axs[0].set_xlabel('saturation [-]')
axs[0].set_ylabel('z-coordinate [m]')
axs[1].set_xlabel('pressure [Pa]')
axs[2].set_xlabel('temperature [K]')
axs[1].ticklabel_format(style='sci', axis='x', scilimits=(0,0))

axs[0].set_ylim([4.5,5])
axs[1].set_ylim([4.5,5])
axs[2].set_ylim([4.5,5])



plt.show()
            
```

Note specifically that the gas saturation in the zoomed-in plot (lower left plot above) indicates that the ice table is at about 4.78 meters, and is therefore below the surface.  This is important -- we don't want to let the ice table fully saturate the subsurface, as it would result in unphysical pressures.  Also, we don't want the ice table to be too far below the surface, as Barrow is a really wet place and the soil is mostly saturated.  Allowing the ice table to be too far below the surface would result in an initial condition that is far out of equilibrium with the site forcing (precipitation, evaporation, runoff data).

Getting the ice table close, but below the surface requires some tweaking of test0 and test2 for new runs.

#### test4: coupled permafrost model testing

To test the permafrost model (which consists of the surface and subsurface, flow and energy equations, we just apply an energy source and start to melt the domain.  

_*This is not a necessary step, and is just done for testing!*_

```python
# plot the solution

def plot(directory, axs, style):
    # load the data
    dati = column_data.column_data(["pressure", "saturation_liquid", 
                                    "saturation_ice", "saturation_gas", "temperature"],
                                    directory=directory)
    keys, times, dat = parse_xmf.readATS(directory=directory)
    skeys, stimes, sdat = parse_xmf.readATS(directory=directory, base="visdump_surface_data.h5")
    
    # convert times to days
    times = [time*365.25 for time in times]

    ice = colors.cm_mapper(-10, times[-1], colors.ice_cmap())
    water = colors.cm_mapper(-10, times[-1], colors.water_cmap())
    gas = colors.cm_mapper(-10, times[-1], colors.gas_cmap())    
    
    # plot saturation of liquid, pressure
    z = dati[0,0,:]
    for i,t in enumerate(times):
        axs[1].plot(dati[5,i,:],z, style, color=gas(t))
        axs[0].plot(dati[2,i,:],z, style, color=water(t))
        axs[0].plot(dati[3,i,:],z, style, color=ice(t))
        axs[0].plot(dati[4,i,:],z, style, color=gas(t))
        
    # inset -- plot surface water
    pd = parse_xmf.getSurfaceData(keys, sdat, 'ponded_depth.cell.0')
    axs[2].plot(times, pd, 'b')
    if pd.max() > 0:
        axs[2].set_ylim([-.1*pd.max(), pd.max()])

    surf_T = parse_xmf.getSurfaceData(keys, sdat, "surface-temperature.cell.0")
    twinax = axs[2].twinx()

    twinax.plot(times, surf_T, 'r')
    twinax.set_ylabel('surface T [K]')

        
        
fig, axs = plt.subplots(1,3, figsize=(12,3))
diii = "test4/test4-v.regression.gold"

plot(diii, axs, '-')

axs[0].set_xlim([-.1,1.1])
axs[0].set_xlabel('saturation [-]')
axs[0].set_ylabel('z-coordinate [m]')
axs[1].set_xlabel('temperature [K]')
axs[2].set_xlabel('time [d]')
axs[2].set_ylabel('ponded depth [m]')

plt.tight_layout()
plt.show()
            
    
# zoom
fig, axs = plt.subplots(1,3, figsize=(12,3))
diii = "test4/test4-v.regression.gold"

plot(diii, axs, '-')

axs[0].set_xlim([-.1,1.1])
axs[0].set_xlabel('saturation [-]')
axs[0].set_ylabel('z-coordinate [m]')
axs[1].set_xlabel('temperature [K]')
axs[2].set_xlabel('time [d]')
axs[2].set_ylabel('ponded depth [m]')

axs[0].set_ylim((4.5,5))
axs[1].set_ylim((4.5,5))

plt.tight_layout()
plt.show()
```

Note the scalloping of the solution in the surface temperature plot -- this is typical and is not an error.  Instead, this is due to the input energy being put into latent heat, not increasing the temperature.  It is, however, a function of the spatial discretization.  If the spatial discretization were finer, this would get smoothed out.  Each scallop corresponds to the next cell down melting.

#### test7: Typical fully coupled transient run

Finally, test7 demonstrates the typical transient column run.  This includes all thermal hydrology processes:

* subsurface flow (Richards)
* subsurface energy
* surface flow
* surface energy
* surface energy balance



```python
# plot the solution

def plot(directory, axs, style):
    # load the data
    dati = column_data.column_data(["pressure", "saturation_liquid", 
                                    "saturation_ice", "saturation_gas", "temperature"],
                                    directory=directory)
    keys, times, dat = parse_xmf.readATS(directory=directory)
    skeys, stimes, sdat = parse_xmf.readATS(directory=directory, base="visdump_surface_data.h5")
    
    # convert times to days
    times = [time*365.25 for time in times]

    ice = colors.cm_mapper(-10, times[-1], colors.ice_cmap())
    water = colors.cm_mapper(-10, times[-1], colors.water_cmap())
    gas = colors.cm_mapper(-10, times[-1], colors.gas_cmap())    
    
    # plot saturation of liquid, pressure
    z = dati[0,0,:]
    for i,t in enumerate(times):
        axs[1].plot(dati[5,i,:],z, style, color=gas(t))
        axs[0].plot(dati[2,i,:],z, style, color=water(t))
        axs[0].plot(dati[3,i,:],z, style, color=ice(t))
        axs[0].plot(dati[4,i,:],z, style, color=gas(t))
        
    # inset -- plot surface water
    pd = parse_xmf.getSurfaceData(keys, sdat, 'ponded_depth.cell.0')
    axs[2].plot(times, pd, 'b')
    if pd.max() > 0:
        axs[2].set_ylim([-.1*pd.max(), pd.max()])

    # plot snow
    snow = parse_xmf.getSurfaceData(keys, sdat, 'snow_depth.cell.0')
    snow = snow + pd
    axs[2].plot(times, snow, 'c')
    if snow.max() > 0:
        axs[2].set_ylim([-.1*snow.max(), snow.max()])
        
    # plot surface temp
    surf_T = parse_xmf.getSurfaceData(keys, sdat, "surface-temperature.cell.0")
    axs[3].plot(times, surf_T, 'r')
    
    # plot snow temp
    snow_T = parse_xmf.getSurfaceData(keys, sdat, 'snow_temperature.cell.0')
    snow_T[0] = np.nan # this is initially zero which screws up the plot
    axs[3].plot(times, snow_T, 'm')

        
fig, axs = plt.subplots(1,3, figsize=(12,3))
twinax = axs[2].twinx()
axs = list(axs)

axs.append(twinax)

diii = "test7/test7-v.regression.gold"
plot(diii, axs, '-')

axs[0].set_xlim([-.1,1.1])
axs[0].set_xlabel('saturation [-]')
axs[0].set_ylabel('z-coordinate [m]')
axs[1].set_xlabel('temperature [K]')
axs[2].set_xlabel('time [d]')
axs[2].set_ylabel('ponded (b) and snow (c) depth [m]')
axs[3].set_ylabel('surface (r) and snow (m) T [K]')

plt.tight_layout()
plt.show()
            
    
# zoom
fig, axs = plt.subplots(1,3, figsize=(12,3))
twinax = axs[2].twinx()
axs = list(axs)
axs.append(twinax)

plot(diii, axs, '-')

axs[0].set_xlim([-.1,1.1])
axs[0].set_xlabel('saturation [-]')
axs[0].set_ylabel('z-coordinate [m]')
axs[1].set_xlabel('temperature [K]')
axs[2].set_xlabel('time [d]')
axs[2].set_ylabel('ponded depth [m]')
axs[3].set_ylabel('surface (r) and snow (m) T [K]')

axs[0].set_ylim((4.5,5))
axs[1].set_ylim((4.5,5))

plt.tight_layout()
plt.show()
```
